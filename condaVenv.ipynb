# How to setup conda virtual env from notebook.

Its best practice to always work from isolated virtual environments with python. This notebook demonstrates how to create such an environment for use with jupyter lab.

### Steps:
1. Create your environment, I named this one here after myself. 
1. Install the ipykernel package into your environment.
1. Activate your environment and install the kernel for jupyter to pick up. 
1. Install a package into that environment, this is how you could install all the dependencies you need. 

That is it! Give Jupyter a few minutes to pick up the new kernel, then you can hit the plus sign, create notebook, and choose the new kernel that is pointing to your virtual env.
